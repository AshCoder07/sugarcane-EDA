```python
import pandas as pd
import seaborn as sns
import numpy as np
from matplotlib import pyplot as plt
```

```python
df=pd.read_csv("List of Countries by Sugarcane Production.csv")
```

```python
df.head()
```

```python
df.shape
```

```python
df.info()
```

## Data Cleaning

```python
df["Production (Tons)"]=df["Production (Tons)"].str.replace(".","")
df["Production per Person (Kg)"]=df["Production per Person (Kg)"].str.replace(".","").str.replace(",",".")
df["Acreage (Hectare)"]=df["Acreage (Hectare)"].str.replace(".","")
df["Yield (Kg / Hectare)"]=df["Yield (Kg / Hectare)"].str.replace(".","").str.replace(",",".")
```

```python
df.head()
```

```python
df.rename(columns={"Production (Tons)":"Production(Tons)"},inplace=True)
df.rename(columns={"Production per Person (Kg)":"Production_per_Person(Kg)"},inplace=True)
df.rename(columns={"Acreage (Hectare)":"Acreage(Hectare)"},inplace=True)
df.rename(columns={"Yield (Kg / Hectare)":"Yield(Kg/Hectare)"},inplace=True)
```

```python
df.head()
```

```python
df.isna().sum()
```

```python
df[df["Acreage(Hectare)"].isna()]
```

```python
df=df.dropna().reset_index()
```

```python
df
```

```python
df.drop(["index","Unnamed: 0"],axis=1,inplace=True)
```

```python
df.head()
```

```python
df.dtypes
```

```python
df["Production(Tons)"]=df["Production(Tons)"].astype("float")
df["Production_per_Person(Kg)"]=df["Production_per_Person(Kg)"].astype("float")
df["Acreage(Hectare)"]=df["Acreage(Hectare)"].astype("float")
df["Yield(Kg/Hectare)"]=df["Yield(Kg/Hectare)"].astype("float")
```

```python
df.dtypes
```

```python
df.nunique()
```

### Which country has the highest production?

```python
df_prodton=df.sort_values("Production(Tons)",ascending=False)
ax=sns.barplot(data=df.head(10),x="Country",y="Production(Tons)")
ax.set_xticklabels(ax.get_xticklabels(),rotation=90)
plt.show()
```

### Which country has highest land?

```python
df_acr=df.sort_values("Acreage(Hectare)",ascending=False)
ax=sns.barplot(data=df_acr.head(10),x="Country",y="Acreage(Hectare)")
ax.set_xticklabels(ax.get_xticklabels(),rotation=90)
plt.show()
```

### Which Country has the highest yield per hectare?

```python
df_yield=df.sort_values("Yield(Kg/Hectare)",ascending=False)
ax=sns.barplot(data=df_yield.head(10),x="Country",y="Yield(Kg/Hectare)")
ax.set_xticklabels(ax.get_xticklabels(),rotation=90)
plt.show()
```

### Which country has the highest production per person?

```python
df_prod=df.sort_values("Production_per_Person(Kg)",ascending=False)
ax=sns.barplot(data=df_prod.head(10),x="Country",y="Production_per_Person(Kg)")
ax.set_xticklabels(ax.get_xticklabels(),rotation=90)
plt.show()
```

## Analysis for Continent

```python
numeric_df = df.select_dtypes(include=[np.number])
numeric_df['Continent'] = df['Continent']
df_cont = numeric_df.groupby("Continent").sum()
```

```python
df_cont
```

### Which Continent produces the maximum sugarcane?

```python
df_cont["Production(Tons)"].sort_values(ascending=False).plot(kind="bar")
```

### Which continent has the highest land?

```python
df_cont["Acreage(Hectare)"].sort_values(ascending=False).plot(kind="bar")
```

### Do number of countries in a continent affect the production?

```python
df_cont["number_of_countries"]=df.groupby("Continent").count()["Country"]
```

```python
df_cont
```

```python
cont_name=df_cont.index.to_list()
sns.lineplot(data=df_cont,x="number_of_countries",y="Production(Tons)")
plt.xticks(df_cont["number_of_countries"],cont_name,rotation=90)
plt.show()
```

### Do Continent with highest land produce more sugarcane?

```python
sns.scatterplot(data=df_cont,x="Acreage(Hectare)",y="Production(Tons)")
```
